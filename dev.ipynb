```python
# Lectura de archivos de configuración
with open('cacsLayoutsConfig.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
# Create empty file
df_test = pd.DataFrame()
print(df_test)
```

```python
# File comlumns
data_columns = df_edos_dia_sga.columns
print(data_columns)
```

```python
# Validate data columns
if 'REASON_VALUE' not in data_columns:
    print("La columna no existe")
```

```python
# Create data file and log file  
for key in keys:
    if key not in data_columns:
        with open("wf_N1_EXT_DAS_CAMBIO_ESTADOS_DIARIOS_FIJA_SGA_log.txt", "w") as log_file:
            log_file.write(f"La columna {key} no existe o esta mal escrita\n")  
        print(f"La columna {key} no existe")
        continue

    df_test[f"{key}"] = df_edos_dia_sga[f"{key}"]
```
